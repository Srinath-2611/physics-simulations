```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
T0 = 2.725  # Present-day temperature of the Universe in Kelvin

# Scale factor range (from early universe to today)
a_vals = np.logspace(-15, 0, 700) # a from 1e-10 to 1

# Temperature as a function of scale factor: T ∝ 1/a
T_vals = T0 / a_vals

# Plotting
plt.figure(figsize=(10, 6))
plt.loglog(a_vals, T_vals, color='darkred')
plt.xlabel('Scale Factor (a)', fontsize=14)
plt.ylabel('Temperature (K)', fontsize=14)
plt.title('Temperature vs. Scale Factor in the Expanding Universe', fontsize=16)
plt.grid(True, which='both', linestyle='--', alpha=0.5)

# Annotate key events (approximate values)
events = {
    'Electroweak Era (~100 GeV)': 1e15,
    'Quark-Hadron Transition (~200 MeV)': 2e12,
    'Big Bang Nucleosynthesis (~1 MeV)': 1e10,
    'Recombination (~3000 K)': 3000,
    'CMB Released (T₀)': T0
}
for label, temp in events.items():
    a_event = T0 / temp
    plt.plot(a_event, temp, 'bo')
    plt.text(a_event * 1.2, temp * 1.2, label, fontsize=10)

plt.tight_layout()
plt.show()
```
