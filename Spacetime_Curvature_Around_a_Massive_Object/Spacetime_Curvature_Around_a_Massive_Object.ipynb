```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
G = 6.67430e-11        # Gravitational constant (arbitrary units)
M = 1000       # Mass of the object
epsilon = 0.1  # Softening factor to avoid singularity at r = 0

# Create a grid of points
grid_size = 100
x = np.linspace(-10, 10, grid_size)
y = np.linspace(-10, 10, grid_size)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2 + epsilon**2)

# Gravitational potential (negative well)
Z = -G * M / R

# Plotting the curved spacetime surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

# Styling
ax.set_title('Spacetime Curvature Around a Massive Object (Analogy)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Curved "Z" dimension (gravitational potential)')
ax.view_init(elev=30, azim=70)  # Adjust for a better view

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set up grid
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# Gravitational potential well (analogy for spacetime curvature)
epsilon = 1e-5  # small value to avoid division by zero
R = np.sqrt(X**2 + Y**2) + epsilon
Z = -1 / R

# Light ray path: simulate bending near the mass (qualitative path)
ray_y = np.linspace(-8, 8, 300)
ray_x = 10 * np.tanh(ray_y / 5)  # simulate deflection
ray_z = -1 / (np.sqrt(ray_x**2 + ray_y**2) + epsilon)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9, edgecolor='none')
ax.plot(ray_x, ray_y, ray_z, color='red', linewidth=2, label='Light Ray')

# Labels and title
ax.set_title('Spacetime Curvature Around a Massive Object (Analogy)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Curved "Z" dimension (gravitational potential)')
ax.legend()

plt.show()
```
