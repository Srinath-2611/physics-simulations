```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
c = 1  # speed of light in natural units (light-years per year)
L = 10  # Distance to the star (in light-years)
v =  0.8 * c  # velocity of traveling twin

# Time for round trip in Earth frame
t_earth = 2 * L / v

# Proper time experienced by traveling twin (dilated)
gamma = 1 / np.sqrt(1 - (v / c)**2)
t_travel = t_earth / gamma

# Time arrays
t1 = np.linspace(0, t_earth, 500)
t2 = np.linspace(0, t_travel, 500)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t1, t1, 'b--', label="Twin A (Stays on Earth)")
plt.plot(np.linspace(0, t_earth, 500), t2, 'r-', label="Twin B (Traveler)")
plt.xlabel("Earth Frame Time (years)")
plt.ylabel("Proper Time Experienced (years)")
plt.title("Twin Paradox Simulation")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
