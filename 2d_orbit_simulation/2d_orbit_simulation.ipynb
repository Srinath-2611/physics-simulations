```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = 6.67430e-11      # Gravitational constant
M = 1.989e30         # Mass of the sun (kg)
dt = 1000            # Time step in seconds
num_steps = 50000    # Number of steps to simulate

# Initial position and velocity of planet (like Earth)
x = [1.496e11]       # 1 AU in meters
y = [0]
vx = [0]
vy = [29780]         # Earthâ€™s orbital speed in m/s
```

```python
for i in range(num_steps):
    r = np.sqrt(x[-1]**2 + y[-1]**2)
    ax = -G * M * x[-1] / r**3
    ay = -G * M * y[-1] / r**3

    vx.append(vx[-1] + ax * dt)
    vy.append(vy[-1] + ay * dt)

    x.append(x[-1] + vx[-1] * dt)
    y.append(y[-1] + vy[-1] * dt)
```

```python
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='blue', label='Planet Orbit')
plt.plot(0, 0, 'o', color='gold', markersize=12, label='Sun')

# Label settings
plt.xlabel("x position (m)", fontsize=12)
plt.ylabel("y position (m)", fontsize=12)
plt.title("2D Planetary Orbit Simulation", fontsize=14, pad=20)

# Add grid, legend, and aspect ratio
plt.grid(True)
plt.legend(loc='upper right')
plt.axis('equal')  # Keeps the orbit circular

# Remove scientific notation on axes for cleaner view (optional)
plt.ticklabel_format(style='plain')

plt.tight_layout()  # Fixes layout padding
plt.show()
```

```python
import numpy as np

G = 6.67430e-11
M = 1.989e30
dt = 1000
num_steps = 5000

x = [1.496e11]
y = [0]
vx = [0]
vy = [29780]

for i in range(num_steps):
    r = np.sqrt(x[-1]**2 + y[-1]**2)
    ax = -G * M * x[-1] / r**3
    ay = -G * M * y[-1] / r**3

    vx.append(vx[-1] + ax * dt)
    vy.append(vy[-1] + ay * dt)

    x.append(x[-1] + vx[-1] * dt)
    y.append(y[-1] + vy[-1] * dt)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-2e11, 2e11)
ax.set_ylim(-2e11, 2e11)
ax.set_aspect('equal')
ax.set_title("Planet Orbiting the Sun")

planet, = ax.plot([], [], 'bo', label="Planet")
sun, = ax.plot(0, 0, 'yo', markersize=12, label="Sun")

def update(i):
    planet.set_data([x[i]], [y[i]])
    return planet,

ani = animation.FuncAnimation(fig, update, frames=len(x), interval=1, blit=True)

plt.legend()
plt.show()
```
