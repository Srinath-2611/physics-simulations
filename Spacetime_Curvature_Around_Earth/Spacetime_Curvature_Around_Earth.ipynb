```python
â€¨import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
G = 6.67430e-11  # m^3 kg^-1 s^-2
c = 3e8          # m/s
M = 5.972e24     # Mass of Earth in kg (for realism)
R = 1e7          # Grid range (10,000 km)

# Schwarzschild radius
Rs = 2 * G * M / c**2  # meters

# Grid
x = np.linspace(-R, R, 300)
y = np.linspace(-R, R, 300)
X, Y = np.meshgrid(x, y)
R_xy = np.sqrt(X**2 + Y**2)

# Gravitational potential Z (realistic scaling using -GM/r)
Z = -G * M / (R_xy + 1e3)  # add small term to avoid division by zero

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X / 1e3, Y / 1e3, Z / 1e7, cmap='viridis')  # scaled for visibility

ax.set_title("Spacetime Curvature Around Earth (Using Real Physical Values)")
ax.set_xlabel("X (km)")
ax.set_ylabel("Y (km)")
ax.set_zlabel("Curved 'Z' dimension (scaled potential)")
plt.tight_layout()
plt.show()
```
