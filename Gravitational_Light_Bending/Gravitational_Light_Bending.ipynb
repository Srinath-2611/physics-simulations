```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11
M = 5e35  # mass of the black hole
c = 3e8   # speed of light
num_rays = 25
y_vals = np.linspace(-1e10, 1e10, num_rays)  # starting y positions

# Simulation settings
dt = 1e3
steps = 3000
```

```python
def simulate_ray(y0):
    x = [ -2e10 ]  # Start left of black hole
    y = [ y0 ]
    vx = [ c ]
    vy = [ 0 ]

    for _ in range(steps):
        r = np.sqrt(x[-1]**2 + y[-1]**2)
        ax = -G * M * x[-1] / r**3
        ay = -G * M * y[-1] / r**3

        vx.append(vx[-1] + ax * dt)
        vy.append(vy[-1] + ay * dt)

        x.append(x[-1] + vx[-1] * dt)
        y.append(y[-1] + vy[-1] * dt)

    return x, y
```

```python
plt.figure(figsize=(8, 8))

# Simulate and plot multiple light rays
for y0 in y_vals:
    x_ray, y_ray = simulate_ray(y0)
    plt.plot(x_ray, y_ray, color='cyan')

# Draw the black hole
plt.plot(0, 0, 'ko', markersize=20, label="Black Hole")
plt.title("Gravitational Light Bending Near a Black Hole")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid(True)
plt.legend()
plt.axis("equal")
plt.show()
```
