```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Time on Earth (in years)
t = np.linspace(0, 20, 500)

# Interactive function
def plot_time_dilation(beta):
    gamma = 1 / np.sqrt(1 - beta**2)
    tau = t / gamma  # Time experienced by the traveler

    plt.figure(figsize=(8, 5))
    plt.plot(t, tau, label=f"Traveling Twin (Î² = {beta:.2f})", color='blue')
    plt.plot(t, t, label="Earth Twin", linestyle='--', color='orange')
    plt.xlabel("Time on Earth (years)")
    plt.ylabel("Experienced Time (years)")
    plt.title("Twin Paradox: Time Dilation with Varying Speed")
    plt.legend()
    plt.grid(True)
    plt.show()

# Slider for beta (speed as a fraction of c)
slider = FloatSlider(min=0.0, max=0.99, step=0.01, value=0.5, description='Î² (v/c):')
interact(plot_time_dilation, beta=slider);
```
